# Games

### Введение
Нам даны исторические данные о продажах игр, оценках пользователей и экспертов, жанрах и платформах (например, Xbox или PlayStation). Необходимо выявить определяющие успешность игры закономерности. Это нужно для того, чтобы интернет-магазин «Стримчик», который продаёт по всему миру компьютерные игры, смог сделать в новом 2017 году  ставку на потенциально популярный продукт и спланировать рекламные кампании. Данные - от самых первых массовых компьютерных и консольных игр и до 2016 года включительно (но данные за 2016 могут быть неполными).


В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию.

Описание данных:

- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board).

## Статистический анализ
Шаг 5. Проверьте гипотезы

    1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
    2. Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Задайте самостоятельно пороговое значение alpha.
Поясните:

    Как вы сформулировали нулевую и альтернативную гипотезы;
    Какой критерий применили для проверки гипотез и почему.

Проверим первое предположение с помощью т-теста для сравнения двух совокупностей. Нулевая гипотеза (H0) будет выглядеть так: 

"Средние пользовательские рейтинги платформ Xbox One и PC одинаковые". 

Посмотрим, сможем ли мы ее опровергнуть. В этом случае придется принять альтернативную гипотезу (H1): 

"Средние пользовательские рейтинги платформ Xbox One и PC не равны". 

Чтобы выполнить задачу, нам нужно сделать срез данных по пользовательским рейтингам и выбрать уровень статистической значимости - возьмем 1%:


```python
x_one = data.query('platform=="XOne" and user_score!=-1')['user_score']
pc = data.query('platform=="PC"and user_score!=-1')['user_score']
```

```python
alpha=0.01

result = st.ttest_ind(x_one,pc,equal_var=False)

print('p-value: ', result.pvalue)

if (result.pvalue < alpha):
        print("Мы отвергаем гипотезу")
else:
        print("Мы не можем отвергнуть гипотезу")
```

Получается, что средние пользовательские рейтинги платформ XOne и PC - одинаковые. Интересно, сколько они составляют и как это равенство выглядит в числах:

```python
print('Средние оценки XOne:',x_one.mean())
print('Средние оценки PC:',pc.mean())
```

Действительно, разница в десятую балла.

Проверим второе предположение. Уровень alpha оставим тот же, 1%. Так как нулевая гипотеза строится обычно на равенстве, переформулируем задачу так:
- Н0: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковы;
- Н1: Средние пользовательские рейтинги жанров Action и Sports не равны.

```python
action = data.query('genre=="Action" and user_score!=-1')['user_score']
sports = data.query('genre=="Sports" and user_score!=-1')['user_score']

alpha=0.01

result = st.ttest_ind(action,sports,equal_var=False)

print('p-value: ', result.pvalue)

if (result.pvalue < alpha):
        print("Мы отвергаем гипотезу")
else:
        print("Мы не можем отвергнуть гипотезу")
```

Здесь результат таков, что мы отвергаем нулевую гипотезу в пользу альтернативной: средние пользовательские рейтинги жанров Action и Sports не равны. Проверим а числах:
